```python
# show the two graph in a one window

import matplotlib.pyplot as plt

student = ['A','B','C','D','E','F']
score = [30,45,47,50,54,59]

plt.subplot(1,2,1)
plt.plot(student,score)
plt.xlabel("student")
plt.ylabel("score")
plt.title("Student score")
plt.grid(True)
plt.show()

plt.subplot(1,2,1)
plt.bar(student,score)
plt.xlabel("student")
plt.ylabel("score")
plt.title("Student score")
plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(10,5))

student = ['A','B','C','D']
score = [30,45,47,50]

# First graph
ax[0].plot(student, score,color="green")
ax[0].set_title('line data')

# Second graph
ax[1].bar(student, score,color= "blue")
ax[1].set_title("Bar data")

fig.suptitle("Compair the line and bar graph")
plt.tight_layout()
plt.show()
```

```python
# import the needed lib for data

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df = df.dropna(subset =['housing_median_age',  'total_rooms',  'total_bedrooms', 'households',  'median_income', ])

households_data = df['households'].value_counts()
plt.figure(figsize=(6,4))
plt.bar(households_data.index ,households_data.values, color='blue')
plt.title('Households Distribution')
plt.xlabel('Households')
plt.ylabel('Count')
plt.savefig('Households.png')
plt.tight_layout()
plt.show()


median_income = df['median_income'].value_counts().sort_index()

plt.figure(figsize=(6,6))
plt.pie(median_income, labels=median_income.index, autopct="%1.1f%%", startangle=90)
plt.xlabel('Median Income')
plt.ylabel('Count')
plt.title('Median Income Distribution')
plt.tight_layout()
plt.savefig('Median_income.png')
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/sample_data/california_housing_train.csv')
df = df.dropna(subset=['housing_median_age', 'total_rooms', 'total_bedrooms', 'households', 'median_income'])

# Group by housing_median_age and calculate average rooms and households
content = df.groupby('housing_median_age')[['total_rooms', 'households']].mean()

plt.figure(figsize=(6,4))
plt.plot(content.index, content['total_rooms'], color='blue', label='Total Rooms')
plt.plot(content.index, content['households'], color='orange', label='Households')

plt.xlabel('Housing Median Age')
plt.ylabel('Value')
plt.title('Comparison of Total Rooms and Households by Housing Median Age')
plt.legend()
plt.tight_layout()
plt.show()
```
